#### Load model from saved weights

```python
checkpoint_path = "training_2/cp-{epoch:04d}.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)
```

```python
latest = tf.train.latest_checkpoint(checkpoint_dir)
latest
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(3)
```

```python
vecs = []
for i in list(trials.keys()):
    vecs.append(trials[i].vec)
vecs = np.array(poly.fit_transform(vecs))
```

```python
trained = makemodel()
trained.load_weights(latest)
```

```python
x = list(trained.predict(vecs))
```

```python
import matplotlib.pyplot as plt
for i in range(len(x)):

    t = x[i]
    if t[0] < 0.5 and i in range(0,20):
        pass
        #print('correct')
    elif t[0] > 0.5 and i in range(20,40):
        pass
        #print('correct')
    else:
        toplt = trials[f'{i+1}'].raw
        plt.plot(toplt['x'],toplt['y'])
        axes = plt.gca()
        axes.set_xlim([0,1500])
        axes.set_ylim([0,500])
        plt.show()
        print(t[0],i)
    #print (i[0])
```
